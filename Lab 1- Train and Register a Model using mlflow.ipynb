```python
from sklearn.model_selection import train_test_split
```

```python
#Load and Prepare the Data
import pandas as pd
url = "https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-01.parquet"
df = pd.read_parquet(url)
```

```python
df
```

```python
#Feature Engineering
df['trip_duration']= (pd.to_datetime(df['tpep_dropoff_datetime']) - pd.to_datetime(df['tpep_pickup_datetime'])).dt.total_seconds() / 60.0
```

```python
df['trip_duration']
```

```python
#filter the trip duration > 0 and < 250 
df= df[(df['trip_duration'] > 0) & (df['trip_duration'] < 250)]
```

```python
df
```

```python
df.columns
```

```python
features= df[['trip_distance', 'trip_duration']]
```

```python
target = df['fare_amount']
```

```python
df['fare_amount'].isna().sum()
```

```python
df['trip_distance'].isna().sum()
```

```python
df['trip_duration'].isna().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30, random_state=42)
```

```python
import mlflow
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mlflow.models import infer_signature
#Start the mlflow context
mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("FareEstimationExperiment")
with mlflow.start_run():
    #Define the model
    model = LinearRegression()
    #Fit the model
    model.fit(X_train, y_train)
    #Predict Fare for the test data
    predictions = model.predict(X_test)
    #Evaluate the model
    mse = mean_squared_error(y_test, predictions)
    #Logging Metrics
    mlflow.log_metric("mse", mse)
    #Logging the hyperparameters
    mlflow.log_param("model_type", "LinearRegression")
    
     # Infer the model signature
    signature = infer_signature(X_train, model.predict(X_train))
    
    # Log the model, which inherits the parameters and metric
    model_info = mlflow.sklearn.log_model(
        sk_model=model,
        name="FareEstimationModel",
        signature=signature,
        input_example=X_train,
        registered_model_name="FareEstimator",
    )
    
   
    
    
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
with mlflow.start_run():
    #Define the model
    gbr = GradientBoostingRegressor(n_estimators=10)
    #Fit the model
    gbr.fit(X_train,y_train)
    #Predict Fare for the test data
    predictions = gbr.predict(X_test)
    #Evaluate the model
    mse = mean_squared_error(y_test, predictions)
    #Logging Metrics
    mlflow.log_metric("mse", mse)
    #Logging the hyperparameters
    mlflow.log_param("model_type", "GradientBoosting Regressor")
    
     # Infer the model signature
    signature = infer_signature(X_train, gbr.predict(X_train))
    
    # Log the model, which inherits the parameters and metric
    model_info = mlflow.sklearn.log_model(
        sk_model=gbr,
        name="FareEstimationModel",
        signature=signature,
        input_example=X_train,
        registered_model_name="FareEstimator",
    )

```

```python
#Loading the registered model with version 1

mlflow.set_tracking_uri("http://localhost:5000")
reg_model_name = 'FareEstimator'
model_uri = f"models:/{reg_model_name}/1"
loaded_model = mlflow.sklearn.load_model(model_uri)
print("Loaded Model",loaded_model)
```
